```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

x=StandardScaler().fit_transform(a)
x=pd.DataFrame(data=x)
x
```

```python
pca=PCA(n_components=70)
princomp=pca.fit_transform(x)
x=pd.DataFrame(data=princomp)

x
```

```python
val_x=x.iloc[40000:60000,:]
val_x.shape
val_y=train_data.iloc[40000:60000,1]
tr_x=x.iloc[0:40000,:]
tr_x.shape
tr_y=train_data.iloc[0:40000,1]
tr_y.shape
type(val_y)
```

```python
from sklearn.ensemble import RandomForestClassifier
reg=RandomForestClassifier(n_estimators=200)
reg.fit(tr_x,tr_y)
pred=reg.predict(val_x)
```

```python
pred=pd.DataFrame(data=pred)
val_y=pd.DataFrame(data=val_y)
print (pred.shape)
print (val_y.shape)
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import f1_score

print(f1_score(val_y,pred))


print(confusion_matrix(val_y,pred))
print(classification_report(val_y,pred))
print(accuracy_score(val_y,pred))
```
